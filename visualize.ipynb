```python
import sys
import numpy as np
sys.path.append("grid_world")
demos = np.load("compile.npy", allow_pickle=True)

import matplotlib.pyplot as plt
from matplotlib.colors import to_rgb

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'thistle', 'tan', 'aquamarine', 'lightgray', 'darkorchid', 'chocolate']
colors = [to_rgb(color) for color in colors]

color_plot = np.array([colors])
plt.imshow(color_plot)

def render(state):
    image = np.zeros((30, 30, 3))
    for i in range(10):
        for j in range(10):
            for k in range(12):
                if state[i][j][k]:
                    if k == 10:
                        image[3*i+1, 3*j+1] = colors[k]
                    else:
                        image[3*i:3*(i+1), 3*j:3*(j+1)] = np.array([[colors[k]]*3]*3)
    return image
```

```python
from utils import add_instruction

fig, axs = plt.subplots(18, 3)
fig.set_size_inches(6, 36)
id = 10
trajectory = add_instruction(demos[id])
for idx, (ax1, ax2, ax3) in enumerate(axs):
    ax1.set_ylim((0, 30))
    ax2.set_ylim((0, 30))
    demo1 = trajectory[idx][0][0]
    demo2 = trajectory[idx][2][0]
    image1 = render(demo1)
    image2 = render(demo2)
    ax1.imshow(image1)
    ax2.imshow(image2)
    ax3.text(0.3, 0.7, str(trajectory[idx][0][1]))
    ax3.text(0.3, 0.5, str(trajectory[idx][1]))
    ax3.text(0.3, 0.3, str(trajectory[idx][2][1]))
plt.show()
```
